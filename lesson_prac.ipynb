```python
## Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

### Importing and exploring the dataset

```python
## iMPORTING THE DATASET
df = pd.read_csv('H:\Machine_Learning\Project\Salary_Data.csv')
df.sample(4)
```

```python
df.columns
```

```python
## Using a heatmap
corr_matrix = df.corr()
sns.heatmap(data=corr_matrix,annot=True,cmap='coolwarm')
```

```python
df.columns
```

```python
## Creating the dependent and independent variables
x = df[['YearsExperience']]
y = df['Salary']
```

```python
## Siplitting the data into train and split data
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
## Training our model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
## Making our predictions
y_pred = model.predict(x_test)

## Displaying the actual and predicted values
predictions = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
predictions
```

```python
## Visualizing to check the linear regression line
plt.scatter(x=x_test,y=y_test,color='green',label='Actual',marker='+')
plt.scatter(x=x_test,y=y_pred,label = "Predicted")

## Plotting the predictions
plt.plot(x_test,y_pred,color='red')

## Creating lagends
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
```

#### Evaluating the linear model

```python
## Mean squar error
mse = mean_squared_error(y_test,y_pred)
print(f"The mean squared error is: {mse}")

# r2 
r2 = r2_score(y_test,y_pred)
print(f"The R-squared score is : {r2}")
```
